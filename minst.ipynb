```python
def accuracy_metrics(y_true, y_predict): 
    from sklearn.metrics import accuracy_score
    accuracy = accuracy_score(y_true, y_predict)
    error_rate = 1 - accuracy
    from sklearn.metrics import precision_score
    precision = precision_score(y_true, y_predict, average='micro')  
    from sklearn.metrics import recall_score
    recall = recall_score(y_true, y_predict, average='micro')  
    return (accuracy, error_rate, precision, recall)
```

```python
from keras.datasets import mnist
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
import matplotlib.pyplot as plt
plt.imshow(X_train[0])
```

```python
#reshape data to fit model
X_train = X_train.reshape(60000,28,28,1)
X_test = X_test.reshape(10000,28,28,1)
```

```python
from keras.utils import to_categorical
#one-hot encode target column
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Conv2D, Flatten
#create model
model = Sequential()
#add model layers
model.add(Conv2D(64, kernel_size=3, activation='relu', input_shape=(28,28,1)))
model.add(Conv2D(32, kernel_size=3, activation='relu'))
model.add(Flatten())
model.add(Dense(10, activation='softmax'))
```

```python
#compile model using accuracy to measure model performance
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
#train the model
model.fit(X_train, y_train, epochs=10)
```

```python
CNNpredict = model.predict(X_test)
```

```python
CNNpredict = CNNpredict.astype(int)
CNNpredict
```

```python
y_test = y_test.astype(int)
y_test
```

```python
CNNacc, CNNerr, CNNpre, CNNrec = accuracy_metrics(y_test, CNNpredict)
print (CNNacc)
print (CNNerr)
print (CNNpre)
print (CNNrec)
```

```python
#train the model
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=3)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Conv2D, Flatten
from keras.regularizers import l2
#create model
model = Sequential()
#add model layers
model.add(Conv2D(64, kernel_size=3, activation='relu', input_shape=(28,28,1), kernel_regularizer=l2(0.01), bias_regularizer=l2(0.01)))
model.add(Conv2D(32, kernel_size=3, activation='relu', kernel_regularizer=l2(0.01), bias_regularizer=l2(0.01)))
model.add(Flatten())
model.add(Dense(10, activation='softmax', kernel_regularizer=l2(0.01), bias_regularizer=l2(0.01)))
```

```python
#compile model using accuracy to measure model performance
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
#train the model
model.fit(X_train, y_train, epochs=3)
```
