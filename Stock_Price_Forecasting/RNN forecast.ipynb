```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset_train = pd.read_csv("/Users/habibabassem/Sypron_Internship_Python_Files/Stock_Price_Forecasting/Google_Stock_Price_Train.csv")
training_set = dataset_train.iloc[:,1:2].values

```

```python
dataset_train
```

```python
training_set
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range = (0,1))
training_set_scaled = sc.fit_transform(training_set)
```

```python
X_train = []
y_train = []
y_train_true = []
for i in range (60, len(training_set_scaled)):
    X_train.append(training_set_scaled[i-60:i, 0])
    y_train.append(training_set_scaled[i,0])
    y_train_true.append(training_set[i,0])
X_train, y_train = np.array(X_train), np.array(y_train)
```

```python
print(X_train.shape)
```

```python
y_train
```

```python
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
```

```python
X_train.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense 
from keras.layers import LSTM
from keras.layers import Dropout
```

```python
regressor = Sequential()
regressor.add(LSTM(units = 50, return_sequences = True, input_shape = (X_train.shape[1],1)))
regressor.add(Dropout(rate = 0.2))

regressor.add(LSTM(units = 50, return_sequences = True))
regressor.add(Dropout(rate = 0.2))

regressor.add(LSTM(units= 50, return_sequences = True))
regressor.add(Dropout(rate = 0.2))

regressor.add(LSTM(units = 50))
regressor.add(Dropout(rate = 0.2))

regressor.add(Dense(units = 1))

regressor.compile(optimizer = 'adam', loss = 'mean_squared_error', metrics = ['accuracy'])
```

```python
X_train.shape
```

```python
y_train = np.reshape(y_train, (y_train.shape[0],1))
y_train.shape
```

```python
regressor.fit(X_train, y_train, epochs = 100, batch_size = 32)
```

```python
dataset_test = pd.read_csv("/Users/habibabassem/Sypron_Internship_Python_Files/Stock_Price_Forecasting/Google_Stock_Price_Test.csv")
real_stock_price = dataset_test.iloc[:,1:2].values 
dataset_total = pd.concat((dataset_train['Open'], dataset_test['Open']), axis = 0)
```

```python
inputs = dataset_total[len(dataset_total) - len(dataset_test) - 60:].values
inputs = inputs.reshape(-1,1)
inputs = sc.transform(inputs)
```

```python
X_test = []
y_test = []
for i in range(60,80):
    X_test.append(inputs[i-60:i, 0])
    y_test.append(inputs[i,0])
X_test = np.array(X_test)
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
y_test = np.array(y_test)
```

```python
predicted_stock_price = regressor.predict(X_test)
predicted_stock_price = sc.inverse_transform(predicted_stock_price)
```

```python
predicted_train = regressor.predict(X_train)
predicted_train = sc.inverse_transform(predicted_train)
```

```python
plt.plot(real_stock_price, color = 'red', label = 'Real Google Stock Price')
plt.plot(predicted_stock_price, color = 'blue', label = 'Predicted Google Stock Price')
plt.title('Google Stock Price Prediction')
plt.xlabel('Time')
plt.ylabel('Google Stock Price')
plt.legend()
plt.show()
```
